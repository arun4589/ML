```python
df=pd.DataFrame(data[['text','target']])
df.head()
```

```python
#some of the steps for text preprocessing for NLP task
#1. lowercasing
df['text']=df['text'].str.lower()
df.tail()
```

```python
#2.removing html tags 
import re
def remove_html_tags(text):
    pattern=re.compile('<.*?>')
    return pattern.sub(r'',text)
```

```python
text='<html><p>body</p>'
remove_html_tags(text)
```

```python
df['text']=df['text'].apply(remove_html_tags)
```

```python
#3.removing urls
def remove_url(text):
    pattern= re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'',text)
```

```python
df['text']=df['text'].apply(remove_url)
```

```python
#4.remove punctuation
import time, string
exclude=string.punctuation
```

```python
def remove_punc(text):
    for ch in exclude:
        text=text.replace(ch,'')
    return text    
```

```python
start=time.time()
df['text'].apply(remove_punc)
end=time.time()
print(end-start)
```

```python
#above process will consume more time as we increase length of dataset so the better approach will be
def remove_punc1(text):
    return text.translate(str.maketrans('','',exclude)) #using this function it will use less time to remove
```

```python

df['text'].apply(remove_punc1)
```

```python
#6. removing stopwords
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
def remove_stopw(text):
    new_text=[]
    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x=new_text[:]
    new_text.clear()
    return " ".join(x)
        
```

```python
remove_stopw(df['text'][0])
```
