```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
X,y=load_iris(return_X_y=True)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
   X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gauss=GaussianNB()
```

```python
gauss.fit(X_train,y_train)
```

```python
y_pred=gauss.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encode=OneHotEncoder()
```

```python
encode.fit_transform(df[['sex','smoker','day']])
```

```python
en_df=pd.DataFrame(encode.fit_transform(df[['sex','smoker','day']]).toarray(),columns=encode.get_feature_names_out())
```

```python
en_df
```

```python
merge=pd.concat([df,en_df],axis=1)
```

```python
merge
```

```python
X=merge.drop(['sex','smoker','day','time'],axis=1)
```

```python
X
```

```python
y=df.time
```

```python
X_train, X_test, y_train, y_test = train_test_split(
   X, y, test_size=0.33, random_state=42)
```

```python
gaus=GaussianNB()
```

```python
gaus.fit(X_train,y_train)
```

```python
y_pred=gaus.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
