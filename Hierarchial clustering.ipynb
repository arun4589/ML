```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
```

```python
df=load_iris()
```

```python
dir(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
   df.data, df.target, test_size=0.20, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_scaled_tr=scaler.fit_transform(X_train)
X_sc_test=scaler.transform(X_test)
```

```python
X_scaled_tr
```

```python
X_sc_test
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2) ## n_componenets =2 means it will reduce features from 4 to 2
```

```python
pca_tr=pca.fit_transform(X_scaled_tr)
pca_test=pca.fit_transform(X_sc_test)
```

```python
## now for the value of cluster we plot its dendogram then we check for a longest line for which there is no horizontal intersecting line
##  to plot we dendogram we have scipy library
```

```python
import scipy.cluster.hierarchy as sc
plt.figure(figsize=(20,20))
plt.title("Dendograms")
sc.dendrogram(sc.linkage(pca_tr,method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster=AgglomerativeClustering(n_clusters=2)
```

```python
cluster.fit(pca_tr)
```

```python
##cluster.fit(pca_test)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_tr[:,0],pca_tr[:,1],c=cluster.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
sil_coef=[]
for i in range(2,11):
    aggl=AgglomerativeClustering(n_clusters=i)
    aggl.fit(pca_tr)
    score=silhouette_score(pca_tr,aggl.labels_)
    sil_coef.append(score)
```

```python
sil_coef
```

```python
plt.plot(range(2,11),sil_coef)
```

```python
## from above graph we can see that at k=2 we get highest silhouttee score which implies that there are two clusters in dataset
```
