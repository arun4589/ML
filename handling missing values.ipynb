```python
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
df
```

```python
df.isnull().sum()## tlls about missing values
```

```python
##dropping rows with missing entries
df.dropna()
```

```python
df.dropna().shape ##dropping rows
```

```python
df.dropna(axis=1).shape ##dropping columns
```

```python
## handling mising values by mean
df['age'].isnull().sum()
```

```python
df['age_mean']=df['age'].fillna(df['age'].mean)
```

```python
df[['age_mean','age']]
```

```python
##by median
df['med_age']=df.age.fillna(df.age.median())
df[['age_mean','age','med_age']]
```

```python
##by mode
```

```python
df[df['embarked'].isnull()]
```

```python
df.embarked.mode()[0]
```

```python
df['embarked_mode']=df.embarked.fillna(df.embarked.mode()[0])
df[['age_mean','age','med_age','embarked_mode']]
```
