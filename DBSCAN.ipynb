```python
import pandas as pd
from sklearn.datasets import make_moons
```

```python
X,y=make_moons(n_samples=1000,noise=0.05)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sca=StandardScaler()
```

```python
x_sca=sca.fit_transform(X)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(x_sca[:,0],x_sca[:,1] ,c=y)
```

```python
from sklearn.cluster import DBSCAN
```

```python
db=DBSCAN(eps=0.5)
db.fit(x_sca)
```

```python
db.labels_
```

```python
plt.scatter(x_sca[:,0],x_sca[:,1] ,c=db.labels_)
```
