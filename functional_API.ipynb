```python
# we can use this api for several purpose such as if we want multiple type of output , if we have multiple type of input
```

for multiple output(example code)


```python
from tensorflow.keras.models import Model
```

```python
from keras.layers import *
```

```python
x=Input(shape=(3,))
```

```python
h1=Dense(128,activation='relu')(x)
h2=Dense(64,activation='relu')(h1)
out1=Dense(1,activation='sigmoid',name='out1')(h2)
out2=Dense(3,activation='softmax',name='out2')(h2)
```

```python
model=Model(inputs=x,outputs=[out1,out2])
```

```python
model.summary()
```

```python
from keras.utils import plot_model
plot_model(model,show_shapes=True)
```

for multiple input(example code)

```python
input1=Input(shape=(32,))
input2=Input(shape=(128,))
```

```python
h11=Dense(32,activation='relu')(input1)
h12=Dense(32,activation='relu')(h11)
h13=Dense(32,activation='relu')(h12)
h21=Dense(64,activation='relu')(input2)
h22=Dense(16,activation='relu')(h21)
concat=concatenate([h13,h22])
ho1=Dense(128,activation='relu')(concat)
out=Dense(1,activation='sigmoid')(ho1)
```

```python
model_=Model(inputs=[input1,input2],outputs=out)
```

```python
model_.summary()
```

```python
plot_model(model_,show_shapes=True)
```
